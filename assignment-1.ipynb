# 1. Reading File to Python Enviornment

```python
#import necessory libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("iris.xls")
print(data)
```

# 2. Display Columns in Data Set

```python
print(data.columns)
```

# 3. Calculate mean of Each Column

```python
data_mean = data[['SL', 'SW', 'PL', 'PW']].mean()
print(data_mean)
```

# 4. Checking for null Values

```python
data.isnull().any()
```

# 5. Visualizations 

#       1. Line Chart

```python
data.plot(title = 'Line Chart of iris Data Set')
plt.show()
```

```python
data[['SL', 'SW']].plot(title = 'Line Chart of SL and SW')
plt.show()
```

#   2. Scatter Plot

```python
data.plot(kind = 'scatter', x = 'PL', y = 'SW' , title = 'Scatter plot for SL and SW')
plt.show()
```

#  3. Histogram

```python
data['SL'].plot(kind = 'hist' , title = 'Histogram of SL'  , rwidth=0.9  , color = 'green')
plt.xlabel("SL")
plt.show()
```
